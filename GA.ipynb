```python
# assume constants
d = 3
q = 4
seed = 42
```

### Δ

```python
import numpy as np
rng = np.random.default_rng(seed=seed)
```

```python
delta = rng.standard_normal((d, 1))
print(f'Fixed Vector Δ:\n', delta)
```

### gaussian vectors

```python
G = np.random.randn(d, q)
for i,g in enumerate(G.transpose()):
  print(f'g_{i+1} : {g}')
```

```python
print(f'Gaussian Matrix G:\n', G)
```

### optimized weights

```python
w = np.dot(G.transpose(), delta)/q

print("w_optimized :\n", w)
```
